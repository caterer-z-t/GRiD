```python
df = final_merged_with_diploid.copy()

# --------------------------------------------------
# 1 Define ID columns
# --------------------------------------------------

id_cols = ["NWD_ID_x", "NWD_ID_y", "NWD_ID", "nwd_id"]

# Only keep columns that actually exist
id_cols = [c for c in id_cols if c in df.columns]

print("Using ID columns:", id_cols)


# --------------------------------------------------
# 2 Create unified NWD_ID column
# (first non-null across columns)
# --------------------------------------------------

df["NWD_ID_final"] = (
    df[id_cols]
    .bfill(axis=1)
    .iloc[:, 0]
)


# --------------------------------------------------
# 3 Detect duplicate IDs
# --------------------------------------------------

duplicate_ids = set(
    df.loc[df["NWD_ID_final"].duplicated(keep=False), "NWD_ID_final"]
)

print(f"\nNumber of duplicate NWD_IDs: {len(duplicate_ids)}")


# --------------------------------------------------
# 4 QC counters
# --------------------------------------------------

num_missing_all_ids = 0
num_id_mismatch = 0
num_missing_lpa = 0
num_missing_cnv = 0
num_duplicate_rows = 0
num_valid_rows = 0

preview_limit = 5
preview_count = 0


# --------------------------------------------------
# 5 Row-wise QC checks
# --------------------------------------------------

for index, row in df.iterrows():

    # ----- Check: at least one ID exists -----
    if all(pd.isna(row[c]) for c in id_cols):
        print(f"\nRow {index} has NO NWD_ID values:")
        print(row[id_cols])
        num_missing_all_ids += 1
        continue

    # ----- Check: ID consistency -----
    non_null_ids = [row[c] for c in id_cols if pd.notna(row[c])]
    
    if len(set(non_null_ids)) > 1:
        print(f"\nRow {index} has mismatched NWD_ID values:")
        print(row[id_cols])
        num_id_mismatch += 1

    # ----- Check: duplicate IDs -----
    if row["NWD_ID_final"] in duplicate_ids:
        print(f"\nRow {index} has duplicate NWD_ID:")
        print(row[id_cols + ["NWD_ID_final"]])
        num_duplicate_rows += 1

    # ----- Check: lpa_nmol -----
    if pd.isna(row["lpa_nmol"]):
        print(f"\nRow {index} missing lpa_nmol:")
        print(row[id_cols + ["lpa_nmol"]])
        num_missing_lpa += 1

    # ----- Check: CNV column -----
    if pd.isna(row["chr6:160605062-160647661"]):
        print(f"\nRow {index} missing chr6 CNV:")
        print(row[id_cols + ["chr6:160605062-160647661"]])
        num_missing_cnv += 1

    # ----- Preview valid rows -----
    if (
        pd.notna(row["NWD_ID_final"])
        and pd.notna(row["lpa_nmol"])
        and pd.notna(row["chr6:160605062-160647661"])
    ):
        num_valid_rows += 1

        if preview_count < preview_limit:
            print(f"\nValid row example (index {index}):")
            print(row[["NWD_ID_final","chr6:160605062-160647661","lpa_nmol"]])
            preview_count += 1


# --------------------------------------------------
# 6 Summary
# --------------------------------------------------

print("\n========== QC SUMMARY ==========")
print("Total rows:", len(df))
print("Rows missing ALL IDs:", num_missing_all_ids)
print("Rows with mismatched IDs:", num_id_mismatch)
print("Rows with duplicate IDs:", num_duplicate_rows)
print("Rows missing lpa_nmol:", num_missing_lpa)
print("Rows missing CNV:", num_missing_cnv)
print("Valid rows:", num_valid_rows)
```

```python
import pandas as pd
import numpy as np

############################################
# Helper function: normalize IDs
############################################

def normalize_id(series):
    return (
        series.astype(str)
        .str.strip()
        .str.upper()
    )

############################################
# Import files
############################################

cross_walk = pd.read_csv(
    "/nas/longleaf/home/catererz/epi/GRiD/files/SOL_NWD_IDs_crosswalk.csv"
)

dip_calls = pd.read_csv(
    "/nas/longleaf/home/catererz/epi/GRiD/files/diploid_calls.txt",
    sep=","
)

anschutz_lpa_subject = pd.read_csv(
    "./files/anschutz_lpa_subject.tsv",
    sep="\t"
)

count_reads_tsv = pd.read_csv(
    "/nas/longleaf/home/catererz/work/updated_LPA_VNTR_SOFTWARE/count-reads.tsv",
    sep="\t"
)

############################################
# Drop unused columns
############################################

cross_walk = cross_walk.drop(columns=[
    "STRAT", "PSU_ID", "GOLD_ID", "GWAS_ID", "HH_ID",
    "SOURCE", "PC1", "PC2", "PC3", "PC4", "PC5",
    "PC6", "PC7", "PC8", "PC9", "PC10", "PC11", "goldID"
])

anschutz_lpa_subject = anschutz_lpa_subject.drop(columns=[
    "entity:subject_id", "pfb:consent_codes", "pfb:created_datetime",
    "pfb:derived_parent_subject_id", "pfb:derived_topmed_subject_id",
    "pfb:project_id", "pfb:state", "pfb:study", "pfb:submitter_id",
    "pfb:updated_datetime"
])

############################################
# Create NWD_ID in count_reads
############################################

count_reads_tsv["NWD_ID"] = (
    count_reads_tsv["Sample"]
    .astype(str)
    .str.split(".").str[0]
)

############################################
# Normalize ALL key IDs
############################################

cross_walk["ID"] = normalize_id(cross_walk["ID"])
dip_calls["id"] = normalize_id(dip_calls["id"])

cross_walk["NWD_ID"] = normalize_id(cross_walk["NWD_ID"])
dip_calls["nwd_id"] = normalize_id(dip_calls["nwd_id"])
count_reads_tsv["NWD_ID"] = normalize_id(count_reads_tsv["NWD_ID"])

############################################
# Merge crosswalk + diploid (LEFT JOIN)
############################################

merged = cross_walk.merge(
    dip_calls,
    left_on="ID",
    right_on="id",
    how="left"   # IMPORTANT: avoid early sample loss
)

print("After crosswalk + diploid merge:", len(merged))

############################################
# Fill missing SUBJECT_ID + NWD_ID (vectorized)
############################################

if "SUBJECT_ID" in merged.columns and "subject_id" in merged.columns:
    merged["SUBJECT_ID"] = merged["SUBJECT_ID"].fillna(merged["subject_id"])
    merged["subject_id"] = merged["subject_id"].fillna(merged["SUBJECT_ID"])

if "NWD_ID" in merged.columns and "nwd_id" in merged.columns:
    merged["NWD_ID"] = merged["NWD_ID"].fillna(merged["nwd_id"])

############################################
# Clean redundant columns
############################################

cols_to_drop = [c for c in ["id", "subject_id", "nwd_id", "SampleID"] if c in merged.columns]
merged = merged.drop(columns=cols_to_drop)

############################################
# Normalize NWD_ID again after filling
############################################

merged["NWD_ID"] = normalize_id(merged["NWD_ID"])

############################################
# DEBUG: Check overlap BEFORE merging reads
############################################

left_ids = set(merged["NWD_ID"].dropna())
right_ids = set(count_reads_tsv["NWD_ID"].dropna())

print("Unique NWD_ID in merged:", len(left_ids))
print("Unique NWD_ID in reads:", len(right_ids))
print("Intersection:", len(left_ids & right_ids))

############################################
# Merge with read counts (LEFT JOIN FIRST)
############################################

final_merged_df = merged.merge(
    count_reads_tsv,
    on="NWD_ID",
    how="left"   # use LEFT first to diagnose
)

############################################
# Diagnostics
############################################

print("\nFinal Merged DataFrame Columns:")
print(final_merged_df.columns.tolist())

print("\nFinal Merged DataFrame Head:")
print(final_merged_df.head(5))

print("\nTotal rows:", len(final_merged_df))

if "chr6:160605062-160647661" in final_merged_df.columns:
    print("Rows with read counts:",
          final_merged_df["chr6:160605062-160647661"].notna().sum())
```
